```python
import pandas as pd 
df = pd.read_csv('cleaned_data.csv')
df.info()
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='price', y='rating')
plt.title('Price vs Rating')
plt.xlabel('Price')
plt.ylabel('Rating')
plt.show()
```

```python
# Calculate correlation between Price and Rating
correlation = df['price'].corr(df['rating'])
print("Correlation between Price and Rating:", correlation)
```

```python
# Check correlation of all features with Rating
correlation_matrix = df.corr(numeric_only=True)
print(correlation_matrix['rating'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
